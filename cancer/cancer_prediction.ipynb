```python
import sklearn
from sklearn import datasets
from sklearn import svm
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
# Set svg format
%matplotlib inline
import matplotlib_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
cancer = datasets.load_breast_cancer()
cancer.feature_names, cancer.target_names
```

```python
x = cancer.data
y = cancer.target
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size = 0.2)
```

```python
clf = svm.SVC(kernel = 'linear', C = 2)
# clf = KNeighborsClassifier(n_neighbors = 9)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
acc = metrics.accuracy_score(y_test, y_pred)
print('Accuracy :', acc)
```

```python
p1 = max(max(y_test), max(y_pred))
p2 = min(min(y_test), min(y_pred))
plt.scatter(y_test, y_pred)
plt.plot([p1, p2], [p1, p2], color = 'red')
plt.show()
```

```python
x_c = np.arange(len(y_pred))
plt.plot(x_c, y_test, color = 'red', label = 'Actual')
plt.plot(x_c, y_pred, color = 'blue', label = 'Prediction')
plt.legend()
plt.show()
```
